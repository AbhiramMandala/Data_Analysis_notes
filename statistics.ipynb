```python
height_updated = Heights - mean_height
print(height_updated)

weight_updated = Weights - mean_weight
print(weight_updated)
```

```python
Mul_height_weight = height_updated * weight_updated
print(Mul_height_weight)
```

```python
popul_cov = sum(Mul_height_weight) / len(Mul_height_weight)
print(popul_cov)
```

```python
Sample_cov =   sum(Mul_height_weight) / (len(Mul_height_weight) - 1)

print(Sample_cov)
```

# Skewness & Kurtosis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate synthetic data
np.random.seed(42)
data = {
    'Normal' : np.random.normal(loc =0 ,scale = 1, size = 1000), # loc : mean , scale : SD
    'Positively_skewed' : np.random.exponential(scale= 2, size =1000 ),  # scale : mean
    'Negatively_skewed' : -np.random.exponential(scale= 2, size =1000 ),  # scale : mean
    'uniform' : np.random.uniform(-3,3,1000)
       }
df = pd.DataFrame(data)
df

print("Skewness")
print(df.skew())
```

# Confidence interval using Z & T tables

```python
import numpy as np
from scipy.stats import norm,t
```

```python
# Confidence Level = 95%, alpha =0.05 , alpha/2 = 0.025 , area = 1 - alpha/2 =1 - 0.025 = 0.975
data = [10, 12, 14, 16, 18]
sqrt_n = np.sqrt(len(data))
x_bar = np.mean(data)

sigma = np.std(data , ddof =0)
z_value = norm.ppf(0.975)

intermediate_val = z_value *( sigma / sqrt_n)

CI_Z = (x_bar  - intermediate_val , x_bar + intermediate_val) 

CI_Z
```

```python
s = np.std(data , ddof =1)

t_value = t.ppf(0.975 , len(data)-1)

intermediate_val2 = t_value *( s / sqrt_n)

CI_T = (x_bar  - intermediate_val2 , x_bar + intermediate_val2) 

CI_T
```

# Two sample t-test(Unpaired test)

```python
# Null hypothesis : In IT industry ,Male Salary is same as Female Salary
# ALternate hypothesis : Male Salary, is different from Female Salary

import numpy as np
from scipy.stats import ttest_ind # 2 Sample ttest function (ind : independent samples)

np.random.seed(42)
male_salary = np.random.randint(20,50 , size = 30) # 
female_salary = np.random.randint(30,60, size = 30)

print("Male :",np.mean(male_salary))
print("Female :",np.mean(female_salary))

#t_statistics , p_value = ttest_ind(male_salary,female_salary) # Two tailed test : male_mean != female_mean
#t_statistics , p_value = ttest_ind(male_salary,female_salary, alternative = 'less')  # Left tailed test : male_mean < female_mean
t_statistics , p_value = ttest_ind(male_salary,female_salary, alternative = 'greater' ) # Right tailed test : male_mean > female_mean


print("t-statistics :", t_statistics)
print("p-value :", p_value)

alpha = 0.05
if p_value <alpha:
    print("Reject Null hypothesis: IT industry is Biased")
else:
    print("Fail to reject Null hypothesis: Not found significant evidence")

# t_statistics variable value : 0, sample mean of male = sample mean of female (approx)
# t_statistics value = -ve : Sample mean of male < Sample mean of female
# t_statistics  value = +ve : Sample mean of male > Sample mean of female
```

# Z -test

## One Sample z-test

```python
import numpy as np
from statsmodels.stats.weightstats import ztest
```

```python
# Null hypothesis  : Sample mean same as population mean 
# Alternate hypothesis : sample mean is signficantly different from population mean

# Example : Null hypothesis : PG canteen owner says his food rating mean is 4.5

sample_ratings = np.array([
                    4.5,3.5,4,4.5,3,2.5,5,4,4.5,2,3,4,1,0,-2,5,3,4,2,3,3.5,3.5,3.8,
                    5,4,1.7,5,5,3,4,2,1.0,-10,-100,0,2,1]
                )

population_mean =4.5
print("Sample mean ",np.mean(sample_ratings))

z_statistics , p_value = ztest(sample_ratings,value = population_mean , alternative= 'two-sided') # Two tailed test
z_statistics , p_value = ztest(sample_ratings,value = population_mean, alternative= 'smaller')  # Left tailed test : sample mean < population mean
z_statistics , p_value = ztest(sample_ratings,value = population_mean, alternative='larger')   # Right tailed test : sample mean > population mean


alpha = 0.04

print("z-statistics :", z_statistics)
print("p-value :", p_value)

if p_value <alpha:
    print("Reject Null hypothesis: PG Canteen owner is lying")
else:
    print("Fail to reject Null hypothesis: Claim is reasonable")

# NOTE : To make ttest inbuilt function works better, we need to provide the large data

# Z_stat = near to 0 : Sample mean = population mean (approx)
# Z_stat = -ve : sample mean < population mean
# Z_stat = +ve : sample mean > population mean
```

# One proportion z-test

```python
##Null hypothesis : Mehfil Resturant claims 90% orders are delivery on time
# Alternate hypothesis : Claim is wrong

from statsmodels.stats.proportion import proportions_ztest

sample_orders = 200
on_time = 50
sample_proportion = on_time / sample_orders
claimed_proportion = 0.90

print("Sample_proportion: ",sample_proportion)

#t_statistics , p_value = proportions_ztest(on_time,sample_orders,value= claimed_proportion)# Two tailed : sample_pro = popul_pro
#t_statistics , p_value = proportions_ztest(on_time,sample_orders,value= claimed_proportion,alternative='smaller')# Left tailed  : sample_pro< popul_pro
t_statistics , p_value = proportions_ztest(on_time,sample_orders,value= claimed_proportion,alternative='larger')# Right tailed : sample_pro > popul_pro


print("t-statistics :", t_statistics)
print("p-value :", p_value)

alpha = 0.05
if p_value <alpha:
    print("Reject Null hypothesis")
else:
    print("Fail to reject Null hypothesis")

# Z_stat = near to 0 : Sample proportion = population proportion (approx)
# Z_stat = -ve : sample proportion < population proportion
# Z_stat = +ve : sample proportion > population proportion
```

```python
# Null hypothesis/Alternate : for two tail,left tail,right tail will be different
# Two tail test:
    H0: sample_proportion = population_proportion
    H1: sample_proportion != population_proportion
# Left tail test:
    H0: sample_proportion >= population_proportion
    H1: sample_proportion < population_proportion
# Right tail test:
    H1: sample_proportion > population_proportion
    H0: sample_proportion <= population_proportion
```

# Two proportion z-test

```python
# NUll hypothesis : proportion of first dataset = proportion of Second dataset
# Alternate hypothesis : proportion of First dataset != proportion of Second dataset

# example:
# Null hypothesis : ON time delivery rate of Swiggy  = On time delivery rate of Zamato
# Alternate hypothesis : There is significant difference between the delivery rate of Swiggy & Zamato

# Swiggy vs Zamato

from statsmodels.stats.proportion import proportions_ztest
counts = [80,60]
sample_size = [100,100]

print("Swiggy sample proportion :",counts[0]/sample_size[0])
print("Zamato sample proportion :",counts[1]/sample_size[1])

z_stat , p_val = proportions_ztest(counts,sample_size)  # Two -tail test
# NUll : Swiggy_proportion >= Zamato_proportion  ,ALter Swiggy_proportion < Zamato_proportion
z_stat , p_val = proportions_ztest(counts,sample_size , alternative= 'smaller')  # Left -tail test
# NUll : Swiggy_proportion <= Zamato_proportion  ,ALter Swiggy_proportion > Zamato_proportion
z_stat , p_val = proportions_ztest(counts,sample_size, alternative= 'larger')  # Right -tail test


print("z-statistics :", z_stat)
print("p-value :", p_val)

alpha = 0.05
if p_val <alpha:
    print("Reject Null hypothesis")
else:
    print("Fail to reject Null hypothesis")
```

# Type -1 & Type - 2 Error

```python
# NUll hypothesis : The coin is fair , p=0.5
# Alternate hypothesis : The coin is not fair (p != 0.5 , p(tail)  != p(head), p(tail) > p(head)) 
import numpy as np
from scipy.stats import binomtest

np.random.seed(42)
n_tosses = 30
alpha = 0.05
n_experiments = 10000

# Finding type - 1 error  : Rejecting a True  NUll hypothesis
# count how many times we are rejecting that coin is fair

false_alarms = 0
for _ in range(n_experiments):
    # Note : binomial() method is used for experiment while binomtest() is used for hypothesis testing
    tosses = np.random.binomial(1, 0.5 , n_tosses)  # doing 30 bermouli trials with p =0.5
    # 1. One bermouli  trial , 0.5 True Probability , n_tosses : 30 bermouli trials per experiment
    heads = np.sum(tosses)

    # p =0.5 ,this is the Null hypothesis probability
    p_value = binomtest(heads , n_tosses , p =0.5 , alternative = 'two-sided')
    if p_value.pvalue > alpha:
        false_alarms += 1 # Count the type -1 error
print(false_alarms)
type1_error_ratio = false_alarms/ n_experiments
print(type1_error_ratio)
```

```python
# NUll hypothesis : The coin is fair , p=0.5
# Alternate hypothesis : The coin is not fair (p != 0.5 , p(tail)  != p(head), p(tail) > p(head)) 
import numpy as np
from scipy.stats import binomtest

np.random.seed(42)
n_tosses = 30
alpha = 0.05
n_experiments = 10000

# Finding type - 2 error  : Failing to Reject A False  NUll hypothesis
# Type 2 Error : If we found p =0.5 (approx) (we find coin is fair) but it is actually p =0.7

false_alarms = 0
for _ in range(n_experiments):
    # Note : binomial() method is used for experiment while binomtest() is used for hypothesis testing
    tosses = np.random.binomial(1, 0.7 , n_tosses)  # doing 30 bermouli trials with p =0.5
    # 1. One bermouli  trial , 0.5 True Probability , n_tosses : 30 bermouli trials per experiment
    heads = np.sum(tosses)

    # p =0.5 ,this is the Null hypothesis probability
    p_value = binomtest(heads , n_tosses , p =0.5 , alternative = 'two-sided')
    if p_value.pvalue > alpha:
        false_alarms += 1 # Count the type -1 error
print(false_alarms)
type2_error_ratio = false_alarms/ n_experiments
print(type2_error_ratio)
```

# Chi - square test

```python

# NUll hypothesis : There is NO significant relationship between two categorical variables
# Alternative hypothesis : There is significant relationship

# Ex: Null hypothesis : There is no  significant relationship between Gender and Buying behaviour

import pandas as pd
from scipy.stats import chi2_contingency

# create  the dataset
#Sales_data = [[10 , 50], [ 50 ,10 ] ]
Sales_data = [[50 , 20], [ 50 ,10 ] ]

df = pd.DataFrame(Sales_data , columns = ["Buy : YES", "Buy : NO"] , index = ["Male", "Female"])
print(df)

chi2_square , p_value , _ , _ =chi2_contingency(df)

# Chi_square : Total mismatch between two categorical variables
print("Chi2_square : ",chi2_square)
print("P_value",p_value)

if p_value < 0.05:
    print("Reject Null hyothesis")
else:
    print("Fail to Reject Null hyothesis")
# There is no two tail or left tail test in chi_square
# Only Right tailed test is exist.
```

# Goodness of fit

```python
# Null hypothesis : Predicted class distribution = Expected class distribution (approx)
# Alternate hypothesis : There is  significant relationship between Predicted & Expected class distribution


from scipy.stats import chisquare
# create the dataset
# rolling the die  60 times

expected_counts = [10,10 , 10, 10 ,10 ,10]
#predicted_counts = [7,8,5,7,3,30]
predicted_counts = [10,10 , 10, 10 ,10 ,10]


chi2_stat , p_value = chisquare(f_obs = predicted_counts , f_exp = expected_counts)

# chi2_stat : Total mismatch between observed and Expected
print("Chi2_square : ",chi2_stat)
print("P_value",p_value)

if p_value < 0.05:
    print("Reject Null hyothesis")
else:
    print("Fail to Reject Null hyothesis")
```
