```python
### Merging two dataframe
### merge() function 

df1=pd.DataFrame({'ID' : [1,2,3] ,'Hero' : ['Allu arjun','Vjiay','Mahesh babu']})
df2=pd.DataFrame({'ID' : [2,3,4] ,'MovieFee' : [100,200,300]})

merged_df  =pd.merge(df1,df2)
# on ='ID'  and how='inner'  ===Default
merged_df
            
```

```python
#Left merge : Take all data of left side table , but only common data from the right side table
merged_df =pd.merge(df1,df2, on='ID',how='left')
merged_df
```

```python
#right merge : Take all data of right side table , but only common data from the leftt side table
merged_df =pd.merge(df1,df2, on='ID',how='right')
merged_df
```

```python
#outer merge : Take all data of right or left side table 
merged_df =pd.merge(df1,df2, on='ID',how='outer')
merged_df
```

```python
## join() function : used to join the dataframe based upon the index
df1= pd.DataFrame({   'Name'  : ['Bill Gates', 'Mark','Elon Musk']} , index=[1,2,3]     )
df2=pd.DataFrame({  'Wealth' : [500,600,800]}  , index=[2,3,4] )

joined_df=df1.join(df2,how='inner')
joined_df
```

```python
joined_df=df1.join(df2,how='left')
joined_df
```

```python
joined_df=df1.join(df2,how='right')
joined_df
```

```python
joined_df=df1.join(df2,how='outer')
joined_df
```

```python
## concat() - combine dataframes (  Vertically(ROW.wise)   or Horizantally(Column Wise))

df1=pd.DataFrame({ 'ID' : [1,2] , 'Movie' : ['Son of India','Kannpa']})
df2=pd.DataFrame({ 'ID' : [3,4] , 'Movie' : ['Mission Impossible','F1']})

concat_df=pd.concat([df1,df2],axis=0,ignore_index=True)
concat_df
```

# 10.Time Series related functions : date_range()

```python
#Generating synthetic time series data
date_rng=pd.date_range(start='2020-01-01',end='2025-12-31',freq='h')
date_rng
```

```python
#selecting 1000 records from above date time data
import numpy as np

np.random.seed(42)    #Randomly given same seed data for you
random_dates=np.random.choice(date_rng,size=1000,replace=False)# Not using same data again
random_dates
```

```python
#Create a synthetic time series Dataset

df=pd.DataFrame( {
    'timestamp' : pd.to_datetime(random_dates),
    'order_id'  : np.arange(1,1001),
    'category'  : np.random.choice(['Biryani','Paya','Irani chai','Maska Bun'],size=1000),
    'Quantity'  : np.random.randint(1,5,size=1000),
    'Price'     : np.random.randint(25,450,size=1000)
        }
            )
df
```

```python
df['Sales']=df['Price'] *  df['Quantity']
```

```python
df=df.sort_values('timestamp').reset_index(drop=True)
```

```python
df['Year']=df.timestamp.dt.year
```

```python
df['Month']=df.timestamp.dt.month
```

```python
df['Day']=df.timestamp.dt.day
```

```python
df['Hour']=df.timestamp.dt.hour
```

```python
df['Minute']=df.timestamp.dt.minute
```

```python
df['Second']=df.timestamp.dt.second
```

```python
df
```
