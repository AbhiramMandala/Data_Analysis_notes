# 8. Concatination, Stacking, Splitting of Numpy array

```python
# concatenate() : This will merge the two or more numpy arrays into one array

# 1D array concatenation
arr1 = np.array([100, 200, 1000])  # pocket money
arr2 = np.array([0, -5, 20]) # remaining pocket money

arr = np.concatenate((arr1, arr2))
print(arr)
```

```python
# 2D arrays concatenate

# axis = 1: Horozontal concatenation ( column wise)

arr1 = np.array([[50, 20, 2],
                 [5, 8, 9]])

arr2 = np.array([[500, 2000, 200],
                 [100, 800, 900]])

arr_2D = np.concatenate((arr1, arr2), axis = 1)
print(arr_2D)
```

```python
# 2D arrays concatenate

# axis = 0: Vertical concatenation ( row-wise)

arr1 = np.array([[50, 20, 2],
                 [5, 8, 9]])

arr2 = np.array([[500, 2000, 200],
                 [100, 800, 900]])

arr_2D = np.concatenate((arr1, arr2), axis = 0)
print(arr_2D)
```

```python
# Interview Question : output of below code?

# IMP Point: We cannot concatenate() two different dimension arrays

a = np.array([1,2,3])
b = np.array([[4,5,6]])

np.concatenate((a,b), axis = 0)  # either put axis = 1
```

```python
# Interview Question : output of below code?

# IMP Point: We can concatenate below arrays row wise

a = np.array([[1,2,3],
              [4,5,6]])  # 2,3
b = np.array([[7,8,9]])  # 1, 3

np.concatenate((a,b), axis = 0) 
```

```python
# Interview Question : output of below code?

# IMP Point: We cannot concatenate below arrays column wise

a = np.array([[1,2,3],
              [4,5,6]])  # 2,3
b = np.array([[7,8,9]])  # 1, 3

np.concatenate((a,b), axis = 1) 
```

### stack() function in numpy

```python
# Def : Stack() function is also used for merging the numpy arrays.
# Stack() function will add one more extra dimension in final output array
```

```python
arr1 = np.array([100, 200, 1000])  # pocket money
arr2 = np.array([0, -5, 20]) # remaining pocket money

# axis = 0:  Row-wise stacking

stacked_arr = np.stack((arr1, arr2), axis = 0)
print(stacked_arr)
```

```python
print(np.concatenate((arr1,arr2)))
```

```python
# axis = 1 : Column wise stacking

stacked_arr = np.stack((arr1, arr2), axis = 1)
print(stacked_arr)
print(stacked_arr.shape)

# n : nmber of input array
# axis = 0 : N, a1,a2......ak
# axis = 1 : N,   
```

```python
 ## vstack() : vartical stacking
# vstack() is same as vartical concatination

arr1 = np.array([[50, 20, 2], [5, 8, 9]])

arr2 = np.array([[500, 2000, 200], [100, 800, 900]])
result = np.vstack((arr1, arr2))
print(result)
```

```python
np.concatenate((arr1, arr2))
```

```python
 ## hstack() : horizontal stacking
# hstack() is same as horizontal concatination

arr1 = np.array([[50, 20, 2], [5, 8, 9]])

arr2 = np.array([[500, 2000, 200], [100, 800, 900]])
result = np.hstack((arr1, arr2))
print(result)
```

```python
np.concatenate((arr1, arr2), axis = 1)
```
