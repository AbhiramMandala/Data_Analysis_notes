```python
import plotly.graph_objects as go
```

# 1.Surface chart : Convex 3D chart

```python
import numpy as np

x= np.linspace(-5,5,30)
y=np.linspace(-5,5,30)
x_new,y_new=np.meshgrid(x,y)  #Converting 1D data to 2D data

z_new = x_new**2 +  y_new**2
#plot the 3D chart
fig=go.Figure(data=[go.Surface(x=x_new,y=y_new,z=z_new)]).update_layout(width=800,height=600)
fig.show()
```

# 2. Surface chart (Non-conex 3D chart)

```python
import numpy as np

x= np.linspace(-5,5,30)
y=np.linspace(-5,5,30)
x_new,y_new=np.meshgrid(x,y)  #Converting 1D data to 2D data

z_new = np.sin(x_new) +  np.cos(y_new)
#plot the 3D chart
fig=go.Figure(data=[go.Surface(x=x_new,y=y_new,z=z_new)]).update_layout(width=800,height=600)
fig.show()
```

# 3. Scatter 3D chart

```python
#Use : find the correlation
np.random.seed(42)

x=np.random.randint(5,20,30)  # Number of chocolates
y=np.random.randint(0,25,30)  # Number of replies
z=np.random.randint(5,30,30)  # Number of Gifts

fig= go.Figure(data=[go.Scatter3d( x = x, y =  y, z = z, mode= "markers")])
fig.update_layout(
    width=800,
    height=600,
    scene= dict(
        xaxis_title ="Number of chocolates",
        yaxis_title = "Number of replies" ,
        zaxis_title =  "Number of Gifts"
    )
)
fig.show()
```

# 4.Multi -Scatter 3D plot

```python
#Use : find the correlation
np.random.seed(42)

# Cluster-1 data : A
x1=np.random.randint(5,20,30)  # Number of chocolates
y1=np.random.randint(0,25,30)  # Number of replies
z1=np.random.randint(5,30,30)  # Number of Gifts

# Cluster-2 data : B
x2=np.random.randint(10,20,30)  # Number of chocolates
y2=np.random.randint(0,25,30)  # Number of replies
z2=np.random.randint(15,30,30)  # Number of Gifts

# Cluster-3 data : C
x3=np.random.randint(10,20,30)  # Number of chocolates
y3=np.random.randint(0,25,30)  # Number of replies
z3=np.random.randint(20,30,30)  # Number of Gifts

fig= go.Figure()
fig.add_trace(go.Scatter3d( x = x1, y =  y1, z = z1, mode= "markers"))
fig.add_trace(go.Scatter3d( x = x2, y =  y2, z = z2, mode= "markers"))
fig.add_trace(go.Scatter3d( x = x3, y =  y3, z = z3, mode= "markers"))

fig.update_layout(
    width=800,
    height=600,
    scene= dict(
        xaxis_title ="Number of chocolates",
        yaxis_title = "Number of replies" ,
        zaxis_title =  "Number of Gifts"
    )
)
fig.show()
```

# 6. 3D Multipline chart

```python
#Use : find the correlation
np.random.seed(42)

days=np.arange(1,31)


# Cluster-1 data : A
x1=days  # Number of chocolates
y1=np.sin(days/3)  # Number of replies
z1=np.cos(days/5)  # Number of Gifts

x2=days  # Number of chocolates
y2=np.sin(days/5)  # Number of replies
z2=np.cos(days/5)  # Number of Gifts

x3=days  # Number of chocolates
y3=np.sin(days/3)  # Number of replies
z3=np.cos(days/3)  # Number of Gifts

fig= go.Figure()
fig.add_trace(go.Scatter3d( x = x1, y =  y1, z = z1, mode="lines",line=dict(width=10)))
fig.add_trace(go.Scatter3d( x = x2, y =  y2, z = z2, mode="lines",line=dict(width=10)))
fig.add_trace(go.Scatter3d( x = x3, y =  y3, z = z3, mode="lines",line=dict(width=10)))

fig.update_layout(
    width=800,
    height=600,
    scene= dict(
        xaxis_title ="Number of chocolates",
        yaxis_title = "Number of replies" ,
        zaxis_title =  "Number of Gifts"
    )
)
fig.show()
```
